# Check Effect of Working Day on Rental Counts

# Null Hypothesis (H0)
The number of electric cycles rented on working days is equal to the number of electric cycles rented on non-working days.

# Alternative Hypothesis (H1)
The number of electric cycles rented on working days is different from the number of electric cycles rented on non-working days.

```python
from scipy.stats import ttest_ind
# Separate data into working and non-working days
workingday_counts = data[data['workingday'] == 1]['count']
non_workingday_counts = data[data['workingday'] == 0]['count']
# Perform 2-sample t-test
t_stat, p_value = ttest_ind(workingday_counts, non_workingday_counts)
# Display the results
t_stat, p_value
```

Assuming the significance level to be 95 percent, thus alpha=0.05 Here value of p is more than
alpha, so v cann't reject the null hypothisis. Thus we can say that working day has no
significant effect on Rental Counts.

Check if Number of Cycles Rented got varied by Season

Null Hypothesis (H0)

The number of electric cycles rented is the same for different seasons.

Alternative Hypothesis (H1)

The number of electric cycles rented is different across different seasons.

```python
from scipy.stats import f_oneway
# Separate data by season
spring_counts = data[data['season'] == 1]['count']
summer_counts = data[data['season'] == 2]['count']
fall_counts = data[data['season'] == 3]['count']
winter_counts = data[data['season'] == 4]['count']
# Perform ANOVA
f_stat, p_value = f_oneway(spring_counts, summer_counts, fall_counts, winter_counts)
# Display the results
f_stat, p_value
```

Assuming the significance level to be 95 percent, the value of p will be significantly low to reject the null
hypothisis,  which means that the seasions have significant effect on Rental Counts.

Check if Number of Cycles Rented Varies by Weather

Null Hypothesis (H0)

The number of electric cycles rented is the same for different weather conditions.

Alternative Hypothesis (H1)

The number of electric cycles rented is different across different weather conditions.

```python
# Separate data by weather
weather_1_counts = data[data['weather'] == 1]['count']
weather_2_counts = data[data['weather'] == 2]['count']
weather_3_counts = data[data['weather'] == 3]['count']
weather_4_counts = data[data['weather'] == 4]['count']
# Perform ANOVA
f_stat, p_value = f_oneway(weather_1_counts, weather_2_counts, weather_3_counts, weather_4_counts)
# Display the results
f_stat, p_value
```

Considering the significance level to be 95 percent, the p value is significantly low to reject the null
hypothisis which means that weather has significant effect on Rental Counts.

Check if Weather is Dependent on Season

Null Hypothesis (H0)
Weather is independent of the season.

Alternative Hypothesis (H1)
Weather is dependent on the season.

```python
from scipy.stats import chi2_contingency
# Create a contingency table
contingency_table = pd.crosstab(data['season'], data['weather'])
# Perform Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
# Display the results
chi2_stat, p_value
```

Considering significance level to be 95 percent, p value is significantly low to reject the null
hypothisis which means that the Weather is dependent on seasion.
